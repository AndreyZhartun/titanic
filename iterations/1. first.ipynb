```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_df =  pd.read_csv('data/train.csv', index_col='PassengerId')
train_df
```

```python
train_df.describe()
```

```python
cat_cols = ['Sex', 'Pclass', 'SibSp', 'Embarked', 'Parch']
[print(train_df[x].value_counts(), '\n') for x in cat_cols]
```

```python
# TODO extract letter from cabin
train_df.Cabin.value_counts()
```

```python
train_df.loc[:, ~train_df.columns.isin(cat_cols)]
```

```python
numeric_cols = ['Age', 'Fare']
train_df[numeric_cols]
```

```python
first_iter_df = train_df[[*cat_cols, *numeric_cols, 'Survived']]
first_iter_df
```

#### Feature Engineering

##### First N/A approach: drop N/A

```python
first_iter_df.Age.notna().sum()
```

```python
first_iter_df = first_iter_df.dropna()
first_iter_df
```

##### Sex feature

```python
if 'Sex' in first_iter_df.columns:
    first_iter_df['is_female'] = (first_iter_df.Sex == 'female').astype('int')
    first_iter_df = first_iter_df.drop(columns=['Sex'])
first_iter_df
```

##### Scaling

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

scaler = StandardScaler()

first_iter_df[numeric_cols] = scaler.fit_transform(first_iter_df[numeric_cols])
first_iter_df
```

```python
cols_to_minmax_scale = cat_cols[:]
cols_to_minmax_scale.remove('Sex')
cols_to_minmax_scale.remove('Embarked')
cols_to_minmax_scale
```

```python
mm_scaler = MinMaxScaler()

first_iter_df[cols_to_minmax_scale] = mm_scaler.fit_transform(first_iter_df[cols_to_minmax_scale])
first_iter_df
```

##### Dummies for Embarked

```python
embarked_dummies = pd.get_dummies(first_iter_df.Embarked, dtype=float)
embarked_dummies
```

```python
first_iter_df['Embarked_C'] = embarked_dummies['C']
first_iter_df['Embarked_S'] = embarked_dummies['S']

if 'Embarked' in first_iter_df.columns:
    first_iter_df = first_iter_df.drop(columns=['Embarked'])

first_iter_df
```

```python
y = first_iter_df.Survived
X = first_iter_df.drop(columns=['Survived'])
train_X = X.to_numpy()
```

```python
train_y = y.to_numpy().reshape(len(y), 1)
```

## Training

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error

kf = KFold(n_splits=5)
```

```python
for i, (train_index, test_index) in enumerate(kf.split(train_X)):
    lr = LogisticRegression()

    fitted = lr.fit(train_X[train_index], train_y[train_index])

    y_pred = fitted.predict(train_X[test_index])
    print(mean_squared_error(train_y[test_index], y_pred))
```
