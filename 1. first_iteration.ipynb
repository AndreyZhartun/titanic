```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_df =  pd.read_csv('data/train.csv', index_col='PassengerId')
train_df
```

First Iteration Preprocessing

```python
from pipeline.feature_preprocessing import preprocess

df = preprocess(train_df)
df
```

```python
y = df.Survived
X = df.drop(columns=['Survived'])
np_X = X.to_numpy()
np_y = y.to_numpy()

print(f'{np_X.shape=}, {np_y.shape=}')
```

```python
from sklearn.linear_model import LogisticRegression, RidgeClassifier
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, accuracy_score

n_splits = 5

kf = KFold(n_splits=n_splits)
losses = np.ndarray(n_splits)
accuracy_list = np.ndarray(n_splits)
```

```python
for i, (train_index, test_index) in enumerate(kf.split(np_X)):
    lr = RidgeClassifier()

    fitted = lr.fit(np_X[train_index], np_y[train_index])

    y_pred = fitted.predict(np_X[test_index])
    y_true = np_y[test_index]
    
    losses[i] = mean_squared_error(y_true, y_pred)
    accuracy_list[i] = accuracy_score(y_true, y_pred)

print(f'{losses=}, {losses.mean()=}')
print(f'{accuracy_list=}, {accuracy_list.mean()=}')
```
